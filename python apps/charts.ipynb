```python
import pandas as pd
```

```python
game_df = pd.read_csv('../resources/game_data.csv')
game_df.head()
```

```python
d = game_df.set_index('name').T.to_dict()
```

```python
game_df = pd.DataFrame.from_dict(d, orient='index')
```

```python
sorted_df = game_df.sort_values(by='ccu', ascending=False).head(10).reset_index()
```

```python
sorted_df
```

```python
top_games = sorted_df['index'].to_list()
concurrent_players = sorted_df['ccu'].to_list()
for i, j in enumerate(top_games):
    print(f'{j}, {concurrent_players[i]}')
```

```python
# Top 10 developers
game_df = pd.DataFrame.from_dict(d, orient='index').reset_index()

a = game_df.groupby('developer').count().sort_values(by='appid', ascending=False).head(10)['index'].index.tolist()
b = [i for i in game_df.groupby('developer').count().sort_values(by='appid', ascending=False).head(10)['index']]

for i, j in enumerate(a):
    print(f'{j}, {b[i]}')
```

```python
a
```

```python
import matplotlib.pyplot as plt

plt.pie(b, labels=a);
```

```python
# Highest rated games
rating_df = game_df[['index', 'positive', 'negative']]
rating_df['rating'] = rating_df['positive'] / (rating_df['positive'] + rating_df['negative'])

a = rating_df.sort_values(by='rating', ascending=False).head(10)['index'].tolist()
b = rating_df.sort_values(by='rating', ascending=False).head(10)['rating'].tolist()

for i, j in enumerate(a):
    print(f'{j}, {round(100 * b[i],2)}%')
```

```python
plt.bar(a, b)
plt.ylim(0.95 * max(b), 1)
```
